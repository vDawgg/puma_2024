```python
import numpy as np
import json
import os
from PIL import Image

im_height, im_width = 1024, 1024

ims_path = "./data/01_training_dataset_tif_ROIs"
ims = []
for im_f in os.listdir(ims_path):
    ims.append(Image.open(os.path.join(ims_path, im_f)))

class_map = {
    "nuclei_tumor": 0,
    "nuclei_lymphocyte": 1,
    "nuclei_plasma_cell": 1,
    "nuclei_histiocyte": 2,
    "nuclei_melanophage": 2,
    "nuclei_neutrophil": 2,
    "nuclei_stroma": 2,
    "nuclei_endothelium": 2,
    "nuclei_epithelium": 2,
    "nuclei_apoptosis": 2,
}

segs_path = "./data/01_training_dataset_geojson_nuclei"
segs = []
for seg_f in os.listdir(segs_path):
    seg_json = json.load(open(os.path.join(segs_path, seg_f)))
    seg = np.zeros((im_height, im_width))
    for nucleus in seg_json["features"]:
        if nucleus["geometry"]["type"] == "MultiPolygon":
            # MultiPolygon: iterate through each polygon's coordinates
            coords = np.array([coord for polygon in nucleus["geometry"]["coordinates"][0] for coord in polygon])
        else:
            # Single Polygon: one list of coordinates
            coords = np.array(nucleus["geometry"]["coordinates"])
        coords = coords.astype(int)
        print(coords)
        print(nucleus["id"])
        print(seg_f)
        for coord in coords:
            seg[coord] = class_map[nucleus["properties"]["classification"]["name"]]
        # seg[coords[:, 0], coords[:, 1]] = class_map[nucleus["properties"]["classification"]["name"]]
    segs.append(seg)
```
